Importing the required Libraries

```python
import cv2

import numpy as np

import matplotlib.pyplot as plt
```

Uploading an Image

```python
def read_file(filename):

    img=cv2.imread(filename)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


    plt.imshow(img)
    plt.show()
    return img

```

Storing an image

```python
originalImage=cv2.imread("/content/CM.jpg")
originalImage=cv2.cvtColor(originalImage,cv2.COLOR_BGR2RGB)
ReSized1=cv2.resize(originalImage,(250,250))
plt.imshow(ReSized1,cmap='gray')
```

Converting Image into grayscale

```python
img = cv2.imread("/content/CM.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
grayScaleImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ReSized2 = cv2.resize(grayScaleImage, (200,200))
plt.imshow(ReSized2, cmap='gray')
```

Smoothening the Image

```python
smoothGrayScale = cv2.medianBlur(grayScaleImage,1)
ReSized3 = cv2.resize(smoothGrayScale, (200, 200))
plt.imshow(ReSized3, cmap='gray')
```

Retrieving the edges of an Image

```python

getEdge = cv2.adaptiveThreshold(smoothGrayScale, 255,
  cv2.ADAPTIVE_THRESH_MEAN_C,
  cv2.THRESH_BINARY, 9, 9)

ReSized4 = cv2.resize(getEdge, (200, 200))
plt.imshow(ReSized4, cmap='gray')
```

Preparing a Mask Image

```python
colorImage = cv2.bilateralFilter(originalImage, 3, 200, 200)
ReSized5 = cv2.resize(colorImage, (200, 200))
plt.imshow(ReSized5, cmap='gray')
```

Giving a Cartoon Effect

```python
cartoonImage = cv2.bitwise_and(colorImage, colorImage, mask=getEdge)

ReSized6 = cv2.resize(cartoonImage, (200, 200))
plt.imshow(ReSized6, cmap='gray')
```

Building all the transitions in a single frame

```python

images=[ReSized1, ReSized2, ReSized3, ReSized4, ReSized5, ReSized6]
fig, axes = plt.subplots(3,2, figsize=(8,8), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')

plt.show()
```
